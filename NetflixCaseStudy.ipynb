```python
pd.DataFrame(data['type'].value_counts())
# to find out how many movies and TV Shows are there
```

```python
pd.DataFrame(data['type'].value_counts(normalize=True)*100)
# to get above in percentage
```

```python
data['type'].value_counts().plot(kind='pie' , autopct='%.2f')
# to visualize the share of movies and TV shows in the data
```

```python
data['type'].value_counts().plot(kind='bar', color=['lightgreen'])
plt.title('Distribution of Content Types')
plt.xlabel('Type')
plt.ylabel('Count')
# to see the distribution
```

```python
popular_cast = data['cast'].str.split(', ').explode().value_counts().head(10)
pd.DataFrame(popular_cast)
#to find the popular cast members:
```

```python
directors = data['director'].value_counts().head(10)
pd.DataFrame(directors)
#to find the most prolific directors:
```

```python
genre_counts = data['listed_in'].str.split(', ').explode().value_counts()
pd.DataFrame(genre_counts)
#To count the frequency of each genre:
```

```python
data['duration_min'] = data['duration'].str.extract('(\d+)').astype(float)
pd.DataFrame(data['duration_min'].describe())
#to extract the duration in minutes and analyze it:
```

```python
top_countries = data['country'].value_counts().head(10)
pd.DataFrame(top_countries)
#to find the top 10 countries with the most content:
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='rating', hue='type', data=data, palette='plasma')
plt.title('Distribution of Content Ratings')
plt.xlabel('Rating')
plt.ylabel('Count')
#to visualize the distribution of content ratings:
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(x='release_year', data=data, palette='magma')
plt.xticks(rotation=90)
plt.title('Number of Content Releases by Year')
plt.show()
#To plot the number of content releases by year:
```

```python
data['date_add']=pd.to_datetime(data['date_added'], errors='coerce')
```

```python
data.info()
```

```python
data.head()
```

```python
data['month']=data['date_add'].dt.month.fillna(-1)
```

```python
data['month']=data['month'].astype('int64')
```

```python
data.head()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='month',hue='type', data=data)
plt.title('Monthly Distribution of Content')
plt.xlabel('Month')
plt.ylabel('Count')
# to find out on which month most movies or TV shows released
```
